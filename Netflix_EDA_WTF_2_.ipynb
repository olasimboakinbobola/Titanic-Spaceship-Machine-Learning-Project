# Cleaning the data for exploration

```python
# importing essential libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the data to be explored
netflix = pd.read_csv('netflix_titles.csv')
netflix
```

```python
# Dropping the column show_id
netflix = netflix.drop(columns= ('show_id'))
netflix.head()
```

```python
# Examining the overview of the dataset
netflix.info()
```

```python
#netflix.describe(include= 'all')
```

```python
netflix.describe()
```

```python
# Checking for missing values in the dataset
netflix.isnull().sum()
```

```python
# dropping the few null rows
netflix = netflix.dropna(subset =['date_added', 'rating', 'duration'])
```

```python
# filling the null rows in the remaining columns
netflix['director'] = netflix['director'].fillna('None Specified')
netflix['cast'] = netflix['cast'].fillna('Not Listed')
netflix['country'] = netflix['country'].fillna('Not Specified')
```

```python
# Checking that all null values have been removed
netflix.isnull().sum()
```

```python
# converting duration to integer
netflix['duration']= netflix['duration'].str.strip('min')
netflix['duration']= netflix['duration'].str.strip('Seasons')
netflix['duration']= netflix['duration'].str.strip('Season')
netflix['duration']= netflix['duration'].str.strip(' ').astype('int')
```

```python
netflix['duration'].value_counts()
```

```python
netflix.head()
```

```python
ratings_meaning= {
    'TV-PG': 'Parental Guidance',
    'TV-MA': 'Mature Audience',
    'TV-Y7-FV': 'Teens',
    'TV-Y7': 'Teens',
    'TV-14': 'Teens',
    'R': 'Mature Audience',
    'TV-Y': 'General Audience',
    'NR': 'Mature Audience',
    'PG-13': 'Teens',
    'TV-G': 'General Audience',
    'PG': 'Teens',
    'G': 'General Audience',
    'UR': 'Mature Audience',
    'NC-17': 'Mature Audience'
      }
american_genre['rating'] = american_genre['rating'].replace(ratings_meaning)
american_genre['rating'].value_counts()
```
